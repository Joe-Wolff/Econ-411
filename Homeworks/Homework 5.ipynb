## 2
    

```python
def total(lst):
    total_ = 0
    n = len(lst)
    for i in range(n):
        total_ += lst[i]
    return total_
total(List)
```

```python
def mean(lst):
    n = len(lst)
    mean_ = total(lst) / n
    return mean_
mean(List)
```

```python
def variance(lst, sample = False):
    list_mean = mean(lst)
    n = len(lst)
    DoF = n - 1
    sum_sq_diff = 0
    for val in lst:
        sum_sq_diff += (val - list_mean) ** 2
    if sample == False:
        variance_ = sum_sq_diff / n
    else:
        variance_ = sum_sq_diff / DoF
    
    return variance_
variance(List, sample= False)
```

```python
import pandas as pd
import numpy as np
import random

macro_dict = {"GDP" : {},
             "Money" : {},
             "Real GDP" : {},
             "Price Level" : {}}

for key in macro_dict:
    for i in range(1990, 2022):
        macro_dict[key][i] = np.random.random() * 10000
        
df=pd.DataFrame(macro_dict)  
df
```
